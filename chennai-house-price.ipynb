# Data Cleaning

```python
#removing all the null values
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.AREA.value_counts()
```

```python
#converting all the values in AREA column to lower case and correcting their spelling.
dataset.AREA=dataset.AREA.str.lower()
dataset.AREA = dataset.AREA.replace({'velchery':'velachery', 
                       'kknagar':'kk nagar', 
                       'tnagar':'t nagar', 
                       'chormpet':'chrompet', 
                       'chrompt':'chrompet', 
                       'chrmpet':'chrompet', 
                       'ana nagar':'anna nagar', 
                       'ann nagar':'anna nagar',
                       'karapakam':'karapakkam', 
                       'adyr':'adyar'})

```

```python
dataset.SALE_COND.value_counts()
```

```python
#converting the SALE_COND feature values into lower case
dataset.SALE_COND=dataset.SALE_COND.str.lower()
#correcting the values in SALE_COND.
dataset.SALE_COND=dataset.SALE_COND.replace({'adj land':'adjland', 
                       'normal sale':'normal sale',
                       'partiall':'partial',
                       'ab normal':'abnormal'})
```

```python
dataset.PARK_FACIL.value_counts()
```

```python
# converting all  PARK_FACIL column values  to lowercase.
dataset.PARK_FACIL = dataset.PARK_FACIL.str.lower()

#correcting the values in PARK_FACIL
dataset.PARK_FACIL=dataset.PARK_FACIL.replace({'noo':'no'})
```

```python
dataset.BUILDTYPE.value_counts()
```

```python
#converting the values in BUILDTYPE feature to lowercase
dataset.BUILDTYPE=dataset.BUILDTYPE.str.lower()

#correcting the values in BUILDTYPE feature
dataset.BUILDTYPE=dataset.BUILDTYPE.replace({'comercial':'commercial', 
                       'others':'other'})
```

```python
dataset.UTILITY_AVAIL.value_counts()
```

```python
#converting the values in UTILITY_AVAIL feature to lowercase.
dataset.UTILITY_AVAIL=dataset.UTILITY_AVAIL.str.lower()

#correcting the values in UTILITY_AVAIL column
dataset.UTILITY_AVAIL=dataset.UTILITY_AVAIL.replace({'all pub':'allpub',
                                             'nosewr':'nosewa'})
```

```python
dataset.STREET.value_counts()
```

```python
#converting the values in STREET feature to lowercase.
dataset.STREET=dataset.STREET.str.lower()

#correcting the values in STREET feature
dataset.STREET=dataset.STREET.replace({'pavd':'paved', 
                       'noaccess':'no access'})
```

```python
#converting the values in MZZONE feature to lowercase.
dataset.MZZONE=dataset.MZZONE.str.lower()
```

```python
#fixing the dtypes of features
dataset.N_BEDROOM=dataset.N_BEDROOM.astype('int')
dataset.N_BATHROOM=dataset.N_BATHROOM.astype('int')

```

```python
dataset.dtypes
```

```python
# Correcting Date columnd from object to daterime format
dataset.DATE_SALE = pd.to_datetime(dataset.DATE_SALE, format='%d-%m-%Y')
dataset.DATE_BUILD = pd.to_datetime(dataset.DATE_BUILD, format='%d-%m-%Y')

# Creating property_age column which deternine how old the property id
dataset['PROPERTY_AGE'] = pd.DatetimeIndex(dataset.DATE_SALE).year - pd.DatetimeIndex(dataset.DATE_BUILD).year
```

# Encoding

```python
#oneHot Encoding
dataset=pd.get_dummies(dataset,columns=['BUILDTYPE'])
```

```python
#DATA MOLDING
# Encoding the area column order-wise 
dataset.AREA = dataset.AREA.map({'karapakkam': 1,
                           'adyar': 2, 
                           'chrompet' : 3,
                           'velachery' : 4,
                           'kk nagar' : 5, 
                           'anna nagar' : 6,
                           't nagar' : 7})

#Encoding Sale_cond
dataset.SALE_COND = dataset.SALE_COND.map({'partial' : 1, 
                                     'family' : 2, 
                                     'abnormal' : 3, 
                                     'normal sale' : 4,
                                     'adjland' : 5})

#Encoding Park_facil
dataset.PARK_FACIL = dataset.PARK_FACIL.map({'yes':1,
                                       'no':0})

#Encoding Utility_avail
dataset.UTILITY_AVAIL = dataset.UTILITY_AVAIL.map({'elo' : 1, 
                                             'nosewa' : 2, 
                                             'nosewr ' : 3,
                                             'allpub' : 4})

#Encoding Street
dataset.STREET = dataset.STREET.map({'no access' : 1,
                               'paved' : 2, 
                               'gravel' : 3})

#Encoding mzzone
dataset.MZZONE = dataset.MZZONE.map({'a' : 1,
                               'c' : 2,
                               'i' : 3,
                               'rl' : 5,
                               'rh' : 4,
                               'rm' : 6})

```

```python
dataset.dtypes
```

```python
# Removing unnecessary columns and Rearranging the columns of dataframe 
dataset = dataset.reindex(columns = ['AREA', 'SALE_COND', 'PARK_FACIL',
       'BUILDTYPE_commercial', 'BUILDTYPE_house','BUILDTYPE_other', 'UTILITY_AVAIL',
       'STREET', 'MZZONE', 'PROPERTY_AGE', 'INT_SQFT', 'DIST_MAINROAD', 'N_BEDROOM',
       'N_BATHROOM', 'N_ROOM', 'QS_ROOMS', 'QS_BATHROOM', 'QS_BEDROOM',
        'QS_OVERALL','SALES_PRICE'])
```

```python
plt.figure(figsize=(18,10), dpi=150)
sns.heatmap(dataset.corr(method='pearson'), cbar=False, annot=True, fmt='.1f', linewidth=0.2, cmap='coolwarm')
```

```python
del dataset["N_ROOM"]
```

```python
#masked data has no impact with the other features so we can delete them
del dataset['QS_ROOMS']
del dataset['QS_BATHROOM']
del dataset['QS_BEDROOM']
del dataset['QS_OVERALL']

#dist mainroad also has no impact so we can delete that feature as well
del dataset['DIST_MAINROAD']
```

```python
X=dataset.drop('SALES_PRICE',axis=1).values
y=dataset.SALES_PRICE.values
```

# splitting into training and test dataset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

# Scaling

```python
from sklearn.preprocessing import RobustScaler
rs=RobustScaler()
X_train=rs.fit_transform(X_train)
X_test=rs.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_sc=sc.fit_transform(X_train)
X_test_sc=sc.transform(X_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
X_train_mms=mms.fit_transform(X_train)
X_test_mms=mms.fit_transform(X_test)
```

INSIGHT:-Here with the help of RobustScaler we are trying to scale the data so that the spread of values between all the features become equal so that the model which train the data by using distance doesn't get very large values as their features.

# Model Training

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr=LinearRegression()
lr.fit(X_train,y_train)

y_pred=lr.predict(X_test)

print('r2_score using RobustScaler :',r2_score(y_test,y_pred))
```

```python
lr.fit(X_train_sc,y_train)

y_pred=lr.predict(X_test_sc)

print('r2_score with StandardScaler :',r2_score(y_test,y_pred))
```

```python
lr.fit(X_train_mms,y_train)

y_pred=lr.predict(X_test_mms)

print('r2_score with MinMaxScaler :',r2_score(y_test,y_pred))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
DT=DecisionTreeRegressor()
DT.fit(X_train,y_train)
y_pred=DT.predict(X_test)
print('r2_score with RobustScaler: ',r2_score(y_test,y_pred))

#with standardscaler
DT.fit(X_train_sc,y_train)
y_pred=DT.predict(X_test_sc)
print('r2_score with standardscaler: ',r2_score(y_test,y_pred))

#with minmaxscaler
DT.fit(X_train_mms,y_train)
y_pred=DT.predict(X_test_mms)
print('r2_score with MinMaxScaler: ',r2_score(y_test,y_pred))
```

# Ensemble Models

# Gradient Boosting 

```python
from sklearn.ensemble import GradientBoostingRegressor
gdr=GradientBoostingRegressor(n_estimators=100,max_depth=50,random_state=0,learning_rate=0.1)
gdr.fit(X_train,y_train)
print('score with robustscaler: ',gdr.score(X_test,y_test))

#with standardscaler
gdr.fit(X_train_sc,y_train)
print('score with standardscaler: ',gdr.score(X_test_sc,y_test))

#with minmaxscaler
gdr.fit(X_train_mms,y_train)
print('score with minmaxscaler: ',gdr.score(X_test_mms,y_test))


```

```python
#Finding the best parameters
from sklearn.model_selection import GridSearchCV
n_est=[100,110,120,150]
max_d=[25,35,45,65]
lr=[0.01,0.05,0.1]

param_grid=dict(n_estimators=n_est,max_depth=max_d,learning_rate=lr)

gb=GradientBoostingRegressor()
grid=GridSearchCV(estimator=gb,param_grid=param_grid,cv=4,n_jobs=-1)
grid_result=grid.fit(X_train,y_train)
print("RobustScaler: ")
print(grid_result.best_params_)
print(grid_result.best_score_)

print("StandardScaler")
grid_result=grid.fit(X_train_sc,y_train)
print(grid_result.best_params_)
print(grid_result.best_score_)

print("MinMaxScaler")
grid_result=grid.fit(X_train_mms,y_train)
print(grid_result.best_params_)
print(grid_result.best_score_)

```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=100,max_depth=15)
rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)
print('r2_score with RobustScaler: ',r2_score(y_test,y_pred))

#standardscaler
rf.fit(X_train_sc,y_train)
y_pred=rf.predict(X_test_sc)
print('r2_score with StandardScaler: ',r2_score(y_test,y_pred))

#minmaxscaler
rf.fit(X_train_mms,y_train)
y_pred=rf.predict(X_test_mms)
print('r2_score with MinMaxScaler: ',r2_score(y_test,y_pred))

```

# XGBoost

```python
import xgboost as xgb
for lr in [0.2,0.5,0.7,0.8,1]:
    model=xgb.XGBRegressor(learning_rate = lr,n_estimators=100)
    model.fit(X_train,y_train)
    print('learning_rate: ',lr,"Training score: ",model.score(X_train,y_train),"Test score: ",model.score(X_test,y_test))
```

```python
print("STANDARDSCALER: ")
for lr in [0.2,0.5,0.7,0.8,1]:
    model=xgb.XGBRegressor(learning_rate = lr,n_estimators=100)
    model.fit(X_train_sc,y_train)
    print('learning_rate: ',lr,"Training score: ",model.score(X_train_sc,y_train),"Test score: ",model.score(X_test_sc,y_test))
```

```python
print("MINMAXSCALER: ")
for lr in [0.2,0.5,0.7,0.8,1]:
    model=xgb.XGBRegressor(learning_rate = lr,n_estimators=100)
    model.fit(X_train_mms,y_train)
    print('learning_rate: ',lr,"Training score: ",model.score(X_train_mms,y_train),"Test score: ",model.score(X_test_mms,y_test))
```

Here after training all kinds of model we can see that MinMaxScaled or StandardScaled data model doesn't matter make much difference they give very similar results.
